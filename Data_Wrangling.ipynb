```python
import pandas as pd
```

```python
df=pd.read_csv("/content/infinixmobile.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
rom=[]
for i in df["Ram"]:
  s=i.split("|")
  rom.append(int(s[1][0:4]))
print(rom)
df["Rom"]=rom
```

```python
ram=[]
for i in df["Ram"]:
  ram.append(int(i[0:2]))
print(ram)
```

```python
df.drop("Ram",inplace=True,axis=1)
df
```

```python
df['Ram']=ram
df.info()
df
```

```python
display=[]
for i in df["Display"]:
  display.append(float(i[0:5]))
print(display)
```

```python
df.drop("Display",inplace=True,axis=1)
df
```

```python
df["Display in cm"]=display
df
```

```python
frontcam=[]
backcam=[]
for i in df['Camera']:
  s=i.split("|")
  if(s[1][2]=='M'):
    frontcam.append(int(s[1][0:2]))
  else:
    frontcam.append(int(s[1][0:3]))
  backcam.append(s[0])
print(frontcam)
print(backcam)

```

```python
df.drop("Camera",inplace=True,axis=1)
df
```

```python
df["Front Camera in Mp"]=frontcam
df["Back Camera in Mp"]=backcam

df
```

```python
price=[]
for i in df["Mobile Price"]:
  s=i.split(",")
  ans=s[0][1:]+s[1]
  price.append(int(ans))
print(price)

```

```python
df.drop("Mobile Price",inplace=True,axis=1)
```

```python
df["MObile Price"]=price
df
```
