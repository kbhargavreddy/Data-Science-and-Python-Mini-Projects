Library management System

functions:
1. add a book
2. return a book
3. Borrow a book
4. print all the books present


```python
# Library Management System

# Initialize library data
library_books = []
borrowed_books = []

# Function to add a book to the library
def add_book(title, author):
    book = {"title": title, "author": author, "borrowed": False}
    library_books.append(book)
    print(f"Book '{title}' by {author} added to the library.")

# Function to borrow a book
def borrow_book(title):
    for book in library_books:
        if book["title"].lower() == title.lower() and not book["borrowed"]:
            book["borrowed"] = True
            borrowed_books.append(book)
            print(f"You have borrowed '{title}'.")
            return
    print(title+" is not available for borrowing")

# Function to return a book
def return_book(title):
    for book in borrowed_books:
        if book["title"].lower() == title.lower():
            book["borrowed"] = False
            borrowed_books.remove(book)
            print(f"You have returned '{title}'. Thank you!")
            return
    print(f"'{title}' is not a borrowed book or not found.")

# Function to print all books
def print_books():
    if not library_books:
        print("The library is empty.")
    else:
        print("Books available in the library:")
        for book in library_books:
            print(f"'{book['title']}' by {book['author']} - {'Available' if not book['borrowed'] else 'Borrowed'}")

# Main program loop
while True:
    print("\nLibrary Management System")
    print("1. Add a book")
    print("2. Borrow a book")
    print("3. Return a book")
    print("4. Print all books")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        title = input("Enter the book title: ")
        author = input("Enter the author name: ")
        add_book(title, author)
    elif choice == '2':
        title = input("Enter the book title you want to borrow: ")
        borrow_book(title)
    elif choice == '3':
        title = input("Enter the book title you want to return: ")
        return_book(title)
    elif choice == '4':
        print_books()
    elif choice == '5':
        print("Thank you for using the Library Management System. Goodbye!")
        break
    else:
        print("Invalid choice. Please select a valid option.")
```
