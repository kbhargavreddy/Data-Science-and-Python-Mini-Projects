Create a datbase with dataframe consisting of 10 players of 2 categories: Bowler and Batsman. Features are 'no of Runs' and 'no of Wickets' achieved. For a batsman, the no of runs > 1000 and wickets < 20. For a bowler, the no of runs < 200 and wickets > 100. Create both scatter plot and bar plots respectively.

```python
import numpy as np
import pandas as pd
data={
    "Name":['ram','shyam','raj','jaggu','sriram','rajesh','bhargav','laksham','dheeraj','saatvik'],
    "runs":[1500,100,2000,120,1768,180,1800,112,4000,102],
    "wickets":[12,120,14,200,11,300,9,250,15,299],
    "type":['batsman','bowler','batsman','bowler','batsman','bowler','batsman','bowler','batsman','bowler']
}
df=pd.DataFrame(data,index=['p1','p2','p3','p4','p5','p6','p7','p8','p9','p10'])
print(df)

import matplotlib.pyplot as plt
x=df['runs']
y=df['Name']
plt.title("runs and names")
plt.scatter(x,y)
plt.show()
x=df['wickets']
y=df['Name']
plt.title("wickets and names")
plt.scatter(x,y,color='black')
plt.show()

c=df['type'].value_counts()
print(c)
plt.bar(c[0],c[1],edgecolor="black",color="red")
plt.show()

x=['bowler','batsman']
y=df['type'].value_counts()
plt.bar(x,y,color='black')
plt.show()
```

```python
df.to_csv('data.csv')
```
